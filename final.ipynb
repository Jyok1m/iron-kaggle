```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Pour afficher les graphiques en ligne dans un notebook Jupyter
%matplotlib inline
```

```python
# Data cleaning
sales_data = pd.read_csv("real_data.csv")
```

```python
sales_df = sales_data
```

```python
# X and y determination
X = sales_df.drop(columns=["index", "date", "open"])

# Preprocessing
from sklearn import preprocessing

converted = preprocessing.LabelEncoder()
converted.fit(['0','a', 'b', 'c', 'd'])
X["state_holiday"] = converted.transform(X["state_holiday"])
```

```python
# Standardization
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
import pickle

ofname = open('linear_model_v2.pkl','rb') #Open in binary format. You never know how it was saved.
linear_model_v2 = pickle.load(ofname)
```

```python
y_pred_sales = linear_model_v2.predict(X)
```

```python
sales_df["sales"] = y_pred_sales
```
